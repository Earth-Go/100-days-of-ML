```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("wine_data.csv" , header = None , usecols = [0,1,2] )
df.columns= ['Class label' , "Alcohol" , "Malic Acid"]
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic Acid'])
```

```python
colours = {1: 'blue' , 3: 'green' , 2: 'red'}
sns.scatterplot(x=df['Alcohol'] , y=df['Malic Acid'], hue = df['Class label'], palette = colours)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
x_train , x_test, y_train, y_test = train_test_split( df.iloc[:, 1:3] , df.iloc[:,0] , test_size=0.3 , random_state= 42 )
x_test
```

```python
x_train
```

```python
Skale = MinMaxScaler()
Skale.fit(x_train)
x_trains = Skale.transform (x_train)
x_tests = Skale.transform(x_test)
x_tests
```

```python
x_trains = pd.DataFrame(x_trains, columns = x_train.columns)
x_tests = pd.DataFrame(x_tests , columns = x_train.columns)
```

```python
np.round(x_train.describe() ,4)
```

```python
np.round(x_trains.describe() ,4)
```

```python
fig, (ax1,ax2) = plt.subplots(ncols =2 , figsize=(12,5))

ax1.scatter(x_trains['Alcohol'] , x_trains['Malic Acid'] , c = y_train)
ax1.set_title("Before Scaling")

ax2.scatter(x_trains['Alcohol'], x_trains['Malic Acid'] , c = y_train)
ax2.set_title("After Scaling")

plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols =2 , figsize=(12,5))

sns.kdeplot(x_train['Alcohol'] , ax =ax1)
sns.kdeplot(x_train['Malic Acid'] ,ax =ax1)
ax1.set_title("Before Scaling")

sns.kdeplot(x_trains['Alcohol'], ax =ax2)
sns.kdeplot( x_trains['Malic Acid'] , ax =ax2)
ax2.set_title("After Scaling")

plt.show()
```

other types of Normalisation involve
i) Mean Normalization :- useful where centered data is required , for that standardization is helpful too! and there is no library involved for it

ii) Mean Absolute :- useful in case of sparse data and has a sklearn library named MaxAbsScaler
                     

Robust Scaling :- usefull when there are a lot of outliers! and is having library RobustScaler!
