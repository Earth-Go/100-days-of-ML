```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier

```

```python
df=pd.read_csv('TitanicDataset.csv')
df
```

```python
df.head()

df.drop(columns = ['PassengerId','Cabin','Name', 'Ticket'], inplace =True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=['Survived']), df['Survived'], test_size =0.2, random_state=42)
```

```python
df.isnull().sum()
x_train.head()
```

```python
print(x_train.head() , y_train.head())
```

```python
#since missing values has to dealt with since they might cause an error in the model prediction system

si_age= SimpleImputer()
si_embarked = SimpleImputer(strategy = 'most_frequent')

x_train_age = si_age.fit_transform(x_train[['Age']])
x_test_age = si_age.transform(x_test[['Age']])

x_train_age = si_embarked.fit_transform(x_train[['Embarked']])
x_test_age = si_embarked.transform(x_test[['Embarked']])
```

```python
ohe_sex = OneHotEncoder(sparse_output=False, handle_unknown = 'ignore')
ohe_embarked = OneHotEncoder(sparse_output=False, handle_unknown = 'ignore')

x_train_sex = ohe_sex.fit_transform(x_train[['Sex']])
x_test_sex = ohe_sex.transform(x_test[['Sex']])

x_train_embarked = ohe_embarked.fit_transform(x_train[['Embarked']])
x_test_embarked = ohe_embarked.transform(x_test[['Embarked']])
```

```python
x_train_rem = x_train.drop(columns = ['Age', 'Sex', 'Embarked'])
x_test_rem = x_test.drop(columns = ['Age', 'Sex', 'Embarked'])
```

```python
print(x_train_rem.head(10))
print(x_test_rem.head(10))
```

```python
x_train_new = np.concatenate((x_train_rem, x_train_age, x_train_sex, x_train_embarked), axis=1)
x_test_new = np.concatenate((x_test_rem, x_test_age, x_test_sex, x_test_embarked), axis=1)
```

```python
x_train_new
```

```python
clf = DecisionTreeClassifier()
#y_train = np.array(y_train)
x_train_new = pd.DataFrame(x_train_new)
x_train_new = x_train_new.apply(pd.to_numeric, errors='coerce')


clf.fit(x_train_new, y_train)
```

```python
x_test_new = pd.DataFrame(x_test_new)
x_test_new = x_test_new.apply(pd.to_numeric, errors='coerce')

y_pred = clf.predict(x_test_new)
```

```python

from sklearn.metrics import accuracy_score
accuracy = accuracy_score( y_pred,y_test )

accuracy
```

```python
import pickle
import os

# Create folder if it doesn't exist
os.makedirs('models', exist_ok=True)

pickle.dump(ohe_sex,open('models/ohe_sex.pkl','wb'))
pickle.dump(ohe_embarked,open('models/ohe_embarked.pkl','wb'))
pickle.dump(clf,open('models/clf.pkl','wb'))
```
